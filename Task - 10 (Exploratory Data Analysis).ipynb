## Removing the outliers

```python
plt.boxplot(df['collegeGPA'])
plt.title('Boxplot before removing the outliers')
plt.show()

def drop_outliers(df, field_name):
    iqr = 1.5 *(np.percentile(df[field_name],75) - np.percentile(df[field_name],25))
    df.drop(df[df[field_name]>(iqr+ np.percentile(df[field_name],75))].index,inplace=True)
    df.drop(df[df[field_name]<(np.percentile(df[field_name],25)-iqr)].index,inplace=True)

drop_outliers(df, 'collegeGPA')
plt.boxplot(df['collegeGPA'])
plt.title('Box plot after removing outliers')
plt.show()
```

## Research question

__Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.__
- - -
Step 1:  
  
Alternate Hypothesis (Bold Claim) :  
$$ H_1: As a fresher if the roles are PA,SE,HE,AE the salary is not equal to 2.5-3.0 lakhs$$ 

Null Hypothesis (status quo):  
$$ H_0: As a fresher if the roles are PA,SE,HE,AE the salary is equal to 2.5-3.0 lakhs $$

step - 2:  
  
- Collect a sample of size n=12  
  
$$ [2.0,3.0,2.7,3.1,1.9,2.2,2.5,2.9,3.0,2.8,2.6,3.1]$$
  
- Compute sample mean  
$$\bar{x}=2.65$$
  
Step - 3: compute Test Statistic:  
  
since standard deviation is not provide we compute t score where,  
$$ t=\frac{\bar{x}-\mu}{\frac{s}{\sqrt[2]{n}}}=-0.744208407535248$$  
Step - 4: Decide $\alpha$ or significance level  
$$\alpha=0.95$$
step - 5.1: Two Tailed t-test:  
$$ Reject\, H_0\, if \, tscore\,  >\, t\, critical $$
$$From\, the \,above\,condition \,we \,failed\,to\,reject\,Null\,Hypothesis$$  
step - 5.2:Calculate P-value  
$$Calculated \,P- value \,is\, :\, 0.4567504024231708 $$
$$if \,P -value > alpha\,: \,reject \,null\, hypothesis$$
$$ Hence,\,we\, failed\, to\, reject\, null\, hypohthesis$$


```python
samples=[2.0,3.0,2.7,3.1,1.9,2.2,2.5,2.9,3.0,2.8,2.6,3.1]
sample_mean=np.array(samples).mean()
print("sample mean:",sample_mean)
def sample_std_(samples,sample_mean):
    summation=0
    for i in range(len(samples)):
        summation=summation+(samples[i]-sample_mean)**2
    
    return (summation/9)**0.5
sample_std=sample_std_(samples,sample_mean)
print("sample standard deviation:",sample_std)
```

```python
from scipy.stats import t
from scipy.stats import norm
def tscore(sample_size,sample_mean,pop_mean,sample_std):
    nume=sample_mean-pop_mean
    denom=sample_std/(sample_size)**0.5
    return nume/denom
```

```python
confidence_level=0.95
alpha=1-confidence_level                                                                                    
t_critical=t.ppf(1-alpha/2,df=9)
print(t_critical)
```

```python
sample_size=12
pop_mean=2.75
```

```python
t_=tscore(sample_size,sample_mean,pop_mean,sample_std)
print(t_)
```

```python
x_min = 1
x_max = 4


# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = sample_std / sample_size**0.5


# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x,mean,std)
plt.xlim(x_min, x_max)
plt.plot(x, y)



# Computing the left and right critical values (Two tailed Test)
t_critical_left = pop_mean + (-t_critical * std)
t_critical_right = pop_mean + (t_critical * std)


# Shading the left rejection region
x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1,mean, std)
plt.fill_between(x1, y1, color='orange')


# Shading the right rejection region
x2 = np.linspace(t_critical_right, x_max, 100)
y2 = norm.pdf(x2,mean,std)
plt.fill_between(x2, y2, color='orange')


# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.0007))
```

```python
if(abs(t_) > t_critical):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
p_value = 2 * (1.0 - norm.cdf(np.abs(t_)))
print("p_value = ", p_value)
if(p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```

- After Performing Hypothesis testing I can conclude that the stats of the Time of India provided in the article is exactly True.Thus,a computer science engineering student could earn 2.5 to 3 lakhs if he performs the roles mentioned in the article.
